# 02 Liability-Assets Flag Check

This notebook verifies whether `Liability-Assets Flag` in the raw dataset is all `1` and shows its full distribution.

```python
import pandas as pd

# Load raw data
df = pd.read_csv('../data/raw/data.csv')
print('Shape:', df.shape)
print('Columns:', len(df.columns))
```

```python
# The dataset columns contain leading spaces, so strip first for robust access
df.columns = [c.strip() for c in df.columns]

col = 'Liability-Assets Flag'
print('Using column:', col)
print('Unique values:', sorted(df[col].dropna().unique().tolist()))
```

```python
# Show exact counts and proportions
counts = df['Liability-Assets Flag'].value_counts(dropna=False).sort_index()
proportions = df['Liability-Assets Flag'].value_counts(dropna=False, normalize=True).sort_index()

summary = pd.DataFrame({
    'count': counts,
    'proportion': proportions
})
summary
```

```python
is_all_one = (df['Liability-Assets Flag'] == 1).all()
print('Is Liability-Assets Flag all 1?', is_all_one)
```
